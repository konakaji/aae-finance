```python
import matplotlib.pyplot as plt

CLASSICAL = "../reports/classical.txt"
ENTROPY = "../reports/quantum.txt"
ENTROPY_IDEAL = "../reports/quantum-ideal.txt"
dates = []
values = []

def plot(file, label):
    dates = []
    values = []
    with open(file) as f:
        for l in f.readlines():
            date, value, _ = l.rstrip().split("\t")
            dates.append(date)
            values.append(float(value))
    plt.plot(dates, values, label=label)

plt.figure(figsize=[12,4])
plt.ylim([0.3, 1.3])
plt.grid()
plot(CLASSICAL, "classical")
plot(ENTROPY, "quantum")
# plot(ENTROPY_IDEAL, "quantum-ideal")

plt.legend()
plt.savefig("../reports/images/entropy.png")
```
