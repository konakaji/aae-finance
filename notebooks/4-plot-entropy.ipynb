```python
import matplotlib.pyplot as plt

CLASSICAL = "../reports/classical.txt"
ENTROPY = "../reports/quantum.txt"
ENTROPY_NAIVE = "../reports/quantum-naive.txt"
ENTROPY_IDEAL = "../reports/quantum-ideal.txt"
dates = []
values = []

def plot(file, label, marker, color, width):
    dates = []
    values = []
    with open(file) as f:
        for l in f.readlines():
            date, value, _ = l.rstrip().split("\t")
            date = date.replace("01", "")
            date = date.replace("02", "")
            date = date.replace("20", "")
            date = date.replace(" ,", "")
            dates.append(date)
            values.append(float(value))
    plt.plot(dates, values, label=label, markersize=10,
             marker=marker, color=color, linewidth=width)

plt.figure(figsize=[20,8])
plt.tick_params(labelsize=20)
plt.xlabel("date", fontsize=20)
plt.ylabel("the SVD Entropy", fontsize=20)
plt.ylim([0.5, 1.5])
plt.title("Transition of the SVD Entropy", fontsize=22)
plt.grid()
plot(CLASSICAL, "exact", color="blue", marker="o", width=2)
plot(ENTROPY, "quantum", color="orange", marker="s", width=3)
plot(ENTROPY_NAIVE, "naive", color="green", marker="x", width=1)

plt.legend(fontsize=20)
plt.savefig("../reports/images/entropy.eps")
```
